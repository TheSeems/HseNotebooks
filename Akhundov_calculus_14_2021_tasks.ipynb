### Задание 1
Изобразить на рисунках область определения функций

1) $z=\ln(x^2+y)$

2) $z=\sqrt{\sin(x^2+y^2)}$

3) $z=\sqrt{x^2-4}+\sqrt{9-y^2}$

```python
x, y, z = symbols('x, y, z')
def contourline(f, x, y, C):
  return sympy.Eq(f(x, y), C)

def draw_2dlvl(func):
  C = symbols('C', real=True)

  p = plot(ylabel='y', show=False)
  for item in [(-2, 'blue'), (-1, 'green'), (0, 'red'), (2, 'black'), (5, 'pink')]:
    c, col = item
    p.append(plot_implicit(contourline(func, x, y, c), line_color=col, line_width=15, show=False)[0])
  
  p.show() 

def draw_3dlvl(func):
  z, c = symbols('z c')
  z12 = solve(sympy.Eq(func, c), z)

  display(*z12)

  с1 = 2
  с2 = 5

  plot3d(z12[0].subs(c, с1), z12[1].subs(c, с1))
  plot3d(z12[0].subs(c, с2), z12[1].subs(c, с2))
```

```python
plot_implicit(x**2 + y > 0)
print()
plot_implicit(Sin(x**2 + y**2) >= 0)
print()
plot_implicit(And(x**2 - 4 >= 0, 9 - y**2 >= 0))
```

### Задача 6.
Проверить выполнение теоремы Эйлера для функции $f(x,y)=\frac{x}{x^2+y^2} $.
###### Указание.
Упростить выражение с помощью simplify, прежде чем сравнивать.

```python
k = symbols('k', real=True)
func = x / (x**2 + y**2)

display('Однородность и измерение', sympy.simplify(func.subs({x: k*x, y: k*y}) / func))
display(sympy.simplify(x*func.diff(x) + y*func.diff(y)), -func)
```
